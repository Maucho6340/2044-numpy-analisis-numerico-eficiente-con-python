### **Desafío: manipulando una base de datos**
Tu misión es:

1.- Crear la base de datos local con la biblioteca SQLAlchemy.

2.- Escribir los datos del archivo CSV en esta base de datos local.
3.- Realizar tres actualizaciones en la base de datos:
  - Actualizar el registro del cliente ID 6840104 cuyo rendimiento anual cambió a 300000.
  - Eliminar el registro de cliente ID 5008809, ya que esta persona ya no tiene cuenta en la institución financiera.
  - Crear un nuevo registro de cliente siguiendo las especificaciones a continuación: ID_cliente: 6850985; Edad: 33; Grado_estudio: Doctorado; Estado_civil: Soltero; Tamaño_familia: 1; Categoria_de_renta: Empleado; Ocupacion: TI; Años_empleado: 2; Rendimiento_anual: 290000; Tiene_carro: 0; Vivienda: Casa/Departamento propio

#### **Consejo importante:**
##### Para agregar un nuevo cliente a la tabla, use la cláusula INSERT INTO seguida del nombre de la tabla y luego especifique los nombres de las columnas de la tabla entre paréntesis. Utilice la cláusula VALUES y luego pase entre paréntesis los nuevos valores para este nuevo registro. Asegúrese de que los valores estén en el orden correcto y en el formato adecuado para cada columna.

```python
# Para crear la base de datos local es necesario ejecutar los siguientes comandos.
# Este código consiste en importar la biblioteca sqlalchemy y crear un objeto engine que
# representa una conexión a una base de datos SQLite en memoria. El método create_engine
# se utiliza para crear el objeto engine y especificar la base de datos que se utilizará.
# En este caso, la base de datos es una base de datos SQLite en memoria, que se crea usando sqlite:///:memory::

import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect, text
engine = create_engine('sqlite:///:memory:')#
```

```python
# A continuación, podemos leer los datos con la función read_csv y escribirlos en el
# formato de la base de datos con la función to_sql:

import pandas as pd
archivo = 'clientes_banco.csv'
datos = pd.read_csv(archivo)
datos.to_sql('clientes', engine, index=False)
```

```python
# Después de crear la tabla, podemos obtener un DataFrame con la función read_sql_table:

pd.read_sql_table('clientes', con=engine.connect())
```

```python
# ¡Ahora vamos a comenzar las actualizaciones de la base de datos!

# Actualizar el registro del cliente ID 6840104 cuyo rendimiento anual cambió a 300000:

# El siguiente código actualiza el rendimiento anual del ID de cliente 6840104 en la tabla clientes.
# La variable query contiene una declaración SQL que actualiza ese registro. Luego, la declaración
# engine.connect() se usa para crear una conexión a la base de datos SQLite. Finalmente,
# la declaración .execute(text(query)) se utiliza para ejecutar la declaración SQL contenida
# en la variable query y actualizar el registro en la tabla clientes.
from sqlalchemy.exc import SQLAlchemyError
query = 'UPDATE clientes SET Rendimiento_anual=300000.0 WHERE ID_Cliente=6840104'
try:
  r_set=engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print("#Registros actualizados: ",r_set.rowcount)
```

```python
# Después de ejecutar este código, puede verificar si la actualización realmente ocurrió leyendo la tabla "clientes"
```

```python
# Eliminar el registro de cliente ID 5008809, ya que esta persona ya no tiene una cuenta en la institución financiera:

from sqlalchemy.exc import SQLAlchemyError
query = 'DELETE FROM clientes WHERE ID_Cliente=5008809'
try:
  r_set=engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print("#Registros borrados: ",r_set.rowcount)
```

###### Cree un nuevo registro de cliente siguiendo las especificaciones dadas anteriormente:

Al trabajar con bases de datos, es común necesitar insertar nuevos registros en nuestras tablas. Para ello podemos utilizar la cláusula INSERT INTO que nos permite añadir nuevos datos de forma práctica y rápida.

Para utilizar esta cláusula, primero debe especificar el nombre de la tabla en la que se insertará el registro. A continuación, debemos pasar entre paréntesis los nombres de las columnas donde se agregarán los nuevos valores. Es importante recordar que el orden de los nombres debe ser el mismo orden en el que se pasarán los valores.

Después de especificar las columnas, usamos la cláusula VALUES para agregar los nuevos valores. Nuevamente debemos pasar entre paréntesis los valores a insertar en el mismo orden en que se especificaron las columnas. Es importante asegurarse de que los valores estén en el formato correcto para cada tipo de datos.

Como la `query` es larga, puedes usar `' \` para indicar que el código continúa en la siguiente línea.

```python
from sqlalchemy.exc import SQLAlchemyError
query = 'INSERT INTO clientes (ID_Cliente, Edad, Grado_estudio, Estado_civil, ' \
        'Tamaño_familia, Categoria_de_renta, Ocupacion, Años_empleado, ' \
        'Rendimiento_anual, Tiene_carro, Vivienda) ' \
        'VALUES (6850985, 33, "Doctorado", "Soltero", 1, "Empleado", "TI", ' \
        '2, 290000, 0, "Casa/Departamento propio")'

try:
  r_set=engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print("#Registros insertados: ",r_set.rowcount)
```

```python
#Para verificar que se realizaron los cambios, puede leer la tabla con el siguiente código:

pd.read_sql_table('clientes', con=engine.connect())
```
